####  5개 단어 문장 삭제 코드 및 2칸 공백 제거 코드 추가
      - 특수 문자 제거 후 2칸 공백이 생길 수 있어, 처리 후 공백제거 코드 추가함

```python
import spacy
import pandas as pd
import re

# spacy
nlp = spacy.load("en_core_web_sm")

# csv 파일
file_path = 'modified_dataset.csv'
df = pd.read_csv(file_path)

# 특수 문자 제거 함수
def remove_special_characters(text):
    # text가 문자열이 아닌 경우 빈 문자열 반환
    if not isinstance(text, str):
        return ""

    # 문자열에서 특수 문자 제거
    pattern = r"[^a-zA-Z0-9+Xx.÷%$₿¢€£¥₩₽ \n]"  # 콤마(,) 제외
    return re.sub(pattern, '', text)

# 연속된 2칸 공백 제거 추가 - 희도님 내용 추가
def replace_multiple_spaces(text):
    return re.sub(' +', ' ', text)

# 단어 개수 세기 함수
def count_words(sentence):
    return len(sentence.split())

# 처리된 데이터를 저장할 리스트 초기화
titles = []
texts = []
sentences = []

# 데이터프레임의 각 행 처리
for i, row in df.iterrows():
    title = remove_special_characters(row['Title'])
    title = replace_multiple_spaces(title)  # 추가된 코드
    text = remove_special_characters(row['Text'])
    text = replace_multiple_spaces(text)  # 추가된 코드

    # spacy로 텍스트 처리하여 문장으로 분리
    doc = nlp(text)

    # spacy Doc 객체에서 문장 추출
    for sent in doc.sents:
        titles.append(title)
        texts.append(text)
        sentences.append(sent.text.lower())

# 결과를 새 데이터프레임에 저장
df_sentences = pd.DataFrame({
    'Title': titles,
    'Original_Text': texts,
    'Sentence': sentences})

# 단어 개수 계산 후 5단어 미만의 문장을 제외
df_sentences['Word_Count'] = df_sentences['Sentence'].apply(count_words)
df_sentences = df_sentences[df_sentences['Word_Count'] >= 5]
df_sentences = df_sentences.drop(columns=['Word_Count'])

# 데이터프레임을 새 csv 파일로 저장
df_sentences.to_csv('concatenated_dataset_1003_v6.csv', index=False)

# 수정된 코드의 결과 출력
df_sentences.head()
```
