```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the data from the uploaded CSV file
file_path = 'result_incorporated.csv'
data = pd.read_csv(file_path)

data.head()  # Displaying the first few rows to understand the structure of the data
```

```python
# Adjusting the plot colors and styles as per the new instructions
column_names = data.columns[1:]



# Defining custom colors and line styles for each train:test ratio
colors = {'2:8': '#E3916B', '5:5': '#8895C2', '8:2': '#BAC263', 'average': '#E6818A'}
line_styles = {'2:8': '-', '5:5': '-', '8:2': '-', 'average': 'solid'}
line_widths = {'2:8': 3, '5:5': 3, '8:2': 3, 'average': 3}  # Making 'average' line thicker

# Creating separate plots for each type of score (acc and f1)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

x_labels = ['No-pt', '100k', '200k', '300k', '400k', '500k']

# Accuracy Scores Plot
for row in data.itertuples():
    axes[0].plot(x_labels, [getattr(row, col) for col in column_names[:6]], label=row[1],
                 color=colors[row[1]], linestyle=line_styles[row[1]], linewidth=line_widths[row[1]], marker='o')
axes[0].set_title('Accuracy', fontsize = 18)
axes[0].legend(title='Train:Test Ratio', loc='lower left', framealpha=0.5, fontsize = 14, title_fontsize=15, frameon = True)
axes[0].set_xticks(x_labels)
axes[0].set_xticklabels(x_labels, fontsize = 15)

# F1 Scores Plot
for row in data.itertuples():
    axes[1].plot(x_labels, [getattr(row, col) for col in column_names[6:]], label=row[1],
                 color=colors[row[1]], linestyle=line_styles[row[1]], linewidth=line_widths[row[1]], marker='o')
axes[1].set_title('F1-Score', fontsize = 18)
axes[1].legend(title='Train:Test Ratio', loc='lower left', framealpha=0.5, fontsize = 14, title_fontsize=15, frameon = True)
axes[1].set_xticks(x_labels)
axes[1].set_xticklabels(x_labels, fontsize = 15)
axes[0].tick_params(axis='y', labelsize=15)
axes[1].tick_params(axis='y', labelsize=15)

plt.tight_layout()
plt.show()
```
